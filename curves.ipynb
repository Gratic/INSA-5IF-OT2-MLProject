```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import json
import os
```

```python
stat_file_path = os.path.abspath("./models/20221130_1134_FifthNeuralNetwork/stats.json")

data = None
with open(stat_file_path, 'r') as f:
    data =json.load(f)
```

```python
data['performances']['train']['losses']
df_loss = pd.DataFrame()
df_accu = pd.DataFrame()
df_loss["train_loss"] = data['performances']['train']['losses']
df_loss["valid_loss"] = data['performances']['valid']['losses']
df_loss["test_loss"] = data['performances']['test']['losses']
df_accu["train_accuracy"] = data['performances']['train']['accuracies']
df_accu["valid_accuracy"] = data['performances']['valid']['accuracies']
df_accu["test_accuracy"] = data['performances']['test']['accuracies']
```

```python
ax = df_accu.plot(figsize=(12, 8), grid=True)
fig = ax.get_figure()
fig.savefig('curves_accuracies.jpg')
```

```python
ax2 = df_loss.plot(figsize=(12, 8), grid=True, logy=True)
fig2 = ax2.get_figure()
fig2.savefig('curves_losses.jpg')
```
